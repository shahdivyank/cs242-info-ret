```python
import numpy as np
```

```python
times = np.load("times.npy")
stopwords = np.load("stopwords.npy")
```

```python
import matplotlib.pyplot as plt

TOTAL_DOCUMENTS = 137375

plt.figure(figsize=(8, 5))

plt.plot(range(TOTAL_DOCUMENTS), times, linestyle='-', label="Naive Indexing")
plt.plot(range(TOTAL_DOCUMENTS), stopwords, linestyle='-', label="Indexing Without Stop Words")


plt.xlabel("Number of Documents")
plt.ylabel("Time (seconds)")
plt.title("Index Creation Performance")

plt.legend()
plt.show()
```
